## 7.3. Visualizing Trends in Number of Graduates Over Time

Description of the Visualization:
X-Axis: Represents the Academic_Year to show the time period.
Y-Axis: Represents the total number of graduates across all graduate categories.
Line Chart: Shows how the number of graduates changes over time, helping to identify upward or downward trends.



```python
# Group the data by 'Academic_Year' and calculate the sum of graduate-related columns
graduate_columns = ['Bachelors', 'Certificate', 'Diploma', 'Doctorate', 
                    'Graduate Certificate', 'Graduate Diploma', 'Higher Diploma', 'Masters']

# Sum all graduate-related columns for each year
graduates_over_time = df.groupby('Academic_Year')[graduate_columns].sum()

# Calculate the total graduates for each year
graduates_over_time['Total Graduates'] = graduates_over_time.sum(axis=1)

# Plot the trend of total graduates over time
plt.figure(figsize=(10, 6))
plt.plot(graduates_over_time.index, graduates_over_time['Total Graduates'], marker='o', color='teal', linewidth=2)

# Add chart title and labels
plt.title('Trends in the Number of Graduates Over Time', fontsize=14)
plt.xlabel('Academic Year', fontsize=12)
plt.ylabel('Total Graduates', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
# If you want to break it down by degree type (e.g., Bachelors, Masters), you can plot multiple lines:
# Plot trends for each degree type
plt.figure(figsize=(10, 6))
for column in graduate_columns:
    plt.plot(graduates_over_time.index, graduates_over_time[column], marker='o', label=column)

# Add chart title and labels
plt.title('Trends in Graduates by Degree Type Over Time', fontsize=14)
plt.xlabel('Academic Year', fontsize=12)
plt.ylabel('Number of Graduates', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Degree Type')
plt.xticks(rotation=45)

# Display the chart
plt.tight_layout()
plt.show()

# Insights You Can Derive:
# Identify which degree type shows the most consistent growth.
# Spot significant increases or decreases for specific years.
# Compare the trends for different degree types.
```

## 7.4. Visualizing Trends in Number of Graduates Over Time

```python
# Group the data by 'Academic_Year' and sum the graduate-related columns
graduate_trends = df.groupby('Academic_Year')[['Bachelors', 'Masters', 'Doctorate']].sum()

# Plot the trends for each graduate type
plt.figure(figsize=(12, 6))
for column in graduate_trends.columns:
    plt.plot(graduate_trends.index, graduate_trends[column], marker='o', label=column)

# Add chart title and labels
plt.title('Trends in the Number of Graduates Over Time', fontsize=16)
plt.xlabel('Academic Year', fontsize=12)
plt.ylabel('Number of Graduates', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Graduate Type', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the chart
plt.show()

# Explanation of Code:
# Data Grouping:
#   We group the dataset by Academic_Year to calculate the total number of graduates for each year.
#   Columns like Bachelors, Masters, and Doctorate are summed to observe trends.
# Line Plot:
#   A separate line is plotted for each graduate type (Bachelors, Masters, Doctorate) using plt.plot().
# Visualization Features:
#   The marker='o' parameter highlights each data point.
#   Labels, gridlines, and legends make the chart easier to read.
```
