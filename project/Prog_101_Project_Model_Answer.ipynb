**3. Apply Data Cleaning Explanation:**

Data cleaning ensures accurate and reliable results. Steps include handling missing values, removing duplicates, and adjusting formats where necessary.

```python
# Drop rows with significant missing values
df.dropna(inplace=True)

# Convert relevant columns to appropriate data types if necessary
df['Year'] = df['Year'].astype(int)

# Check for duplicates and remove them
df.drop_duplicates(inplace=True)

# Display cleaned dataset summary
df.info()
df.head()
```

Explanation: The data cleaning involves removing any rows with missing values to ensure consistency. This is necessary for accurate analysis, especially in statistical and conditional operations.

Question 1: Statistical Analysis

  # 1. What is the average Happiness Score of all countries?

```python
avg_happiness = df['Happiness Score'].mean()
print("Average Happiness Score:", avg_happiness)
```

# 2. How many countries have an Economy score over 1.0 and a Freedom score over 0.5?

```python
high_scores = df[(df['Economy'] > 1.0) & (df['Freedom'] > 0.5)]
print("Number of countries:", len(high_scores))
```

# 3. List the countries with Health scores less than 0.1 and Happiness Scores below 5.

```python
low_happiness_health = df[(df['Health'] < 0.1) & (df['Happiness Score'] < 5)]
low_happiness_health[['Country', 'Happiness Score', 'Health']]
```

# 4. What is the average Happiness Score for each region?

```python
avg_happiness_by_region = df.groupby('Region')['Happiness Score'].mean()
print(avg_happiness_by_region)
```

# 5. List the top 5 happiest countries.

```python
top_happy_countries = df.sort_values(by='Happiness Score', ascending=False).head(5)
top_happy_countries[['Country', 'Happiness Score']]
```

# 6. What are the top 5 countries in the region with the highest average Happiness Score?

```python
# Find the region with the highest average Happiness Score
top_region = avg_happiness_by_region.idxmax()
top_countries_in_region = df[df['Region'] == top_region].sort_values(by='Happiness Score', ascending=False).head(5)
top_countries_in_region[['Country', 'Happiness Score']]
```

# 7. List 6 countries with above-average Economy and Health scores, sorted by Freedom.

```python
avg_economy = df['Economy'].mean()
avg_health = df['Health'].mean()
top_countries = df[(df['Economy'] > avg_economy) & (df['Health'] > avg_health)].sort_values(by='Freedom', ascending=False)
top_countries[['Country', 'Economy', 'Health', 'Freedom']].head(6)
```

Step 4: Visualize the Data

We’ll create some simple charts to make the data easier to understand.

# 8. Show the average Happiness Score in each region (Bar Chart)

```python
import matplotlib.pyplot as plt

avg_happiness_by_region.plot(kind='bar', color='skyblue')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.show()
```

# 9. Show the relationship between Economy and Happiness Score (Scatter Plot)

```python
plt.scatter(df['Economy'], df['Happiness Score'], color='green')
plt.title('Happiness Score vs. Economy')
plt.xlabel('Economy')
plt.ylabel('Happiness Score')
plt.show()
```

# 10. Show the proportion of countries in each region (Pie Chart)

```python
region_counts = df['Region'].value_counts()
region_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Countries by Region')
plt.ylabel('')  # Hide y-label for cleaner look
plt.show()
```
