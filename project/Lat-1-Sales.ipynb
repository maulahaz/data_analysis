```python
#--Check Detail data which having null value:
dt_null_year = pd.isnull(df['Year'])
df[dt_null_year].head(3)
```

```python
#--Make a Function to Check Detail data which having null value:
def check_detail_null(column):
    dt_null = pd.isnull(df[column])
    return df[dt_null].head(3)

check_detail_null('Publisher')
```

```python
#--Handling Null value:
#1. Ignore or Drop if it's not really important
#2. Fill with 0 or Mean Value
#3. Filling a missing value with previous ones. 'df.fillna(method ='pad')'
#4. Filling null value with the next ones. 'df.fillna(method ='bfill')'
#5. Replace the all Nan value in the data frame with -99 value. 'df.replace(to_replace = np.nan, value = -99)' 
#6. Drop a Rows whose having NULL. 'df.dropna(how = 'all')'
#7. Drop a Columns whose having NULL. 'df.dropna(axis = 1)'

##--NOTES--##
# if df.fillna____ => mean FILL ALL NULL VALUE
# if df['Column Name'].fillna___ => mean FILL ONLY at That Column

df.fillna(method ='pad', inplace=True)
df.loc[170:190]
```
